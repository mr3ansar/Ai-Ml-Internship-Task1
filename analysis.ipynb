```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#Loading Iris Dataset from Seaborn
iris_data = sns.load_dataset("iris")
#printing Shape and Showing Head Columns with Column names
shape = iris_data.shape
print(shape)
iris_data.head()
```

```python
#Showing Summary
iris_data.describe()
```

```python
#Showing Info
iris_data.info()
```

```python
#Plotting Scatter Plot for Sepal Length Vs Petal Length
sns.scatterplot(data=iris_data, x = "sepal_length", y = "petal_length", hue="species")
plt.title("Sepal Length Vs Petal Length")
plt.savefig("plots/scatter_plot.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
#Scatter Plot For Relationship Between All Features
iris_data.hist(figsize=(10, 8), edgecolor='black')
plt.tight_layout()
plt.savefig("plots/scatter_plot(all_features).png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
#Ploting Histogram for Sepal Lenght Distribution
iris_data["sepal_length"].hist(bins= 20, edgecolor="black")
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length")
plt.savefig("plots/histogram_plot.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
#Sepal Length By Outliers Using Box Plot
sns.boxplot(data=iris_data, x="species", y="sepal_length")
plt.title("Sepal Length By Species")
plt.savefig("plots/box_plot.png", dpi=150, bbox_inches="tight")
plt.show()
```
